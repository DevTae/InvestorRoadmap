```python
import glob
import pandas as pd

def read_csv_files(directory_path):
    # 지정된 디렉토리에서 모든 CSV 파일 찾기
    csv_files = glob.glob(f'{directory_path}/*.csv')

    # CSV 파일을 저장할 리스트 초기화
    data_frames = []

    # 각 CSV 파일을 읽어서 리스트에 추가
    for csv_file in csv_files:
        try:
            # CSV 파일을 DataFrame으로 읽기
            df = pd.read_csv(csv_file, encoding='euc-kr')
            
            # DataFrame을 리스트에 추가
            data_frames.append(df)
        except pd.errors.EmptyDataError:
            # 빈 CSV 파일의 경우 예외 처리
            print(f"Warning: {csv_file} is empty.")

    return data_frames

# CSV 파일이 있는 디렉토리 경로 지정
directory_path = r"C:\Users\user\OneDrive\바탕 화면\db 공모전\data\3분기\재무상태표"

# CSV 파일들을 읽어와서 리스트에 담기
fs_local_files = read_csv_files(directory_path)

# 리스트에 있는 데이터 프레임들을 확인하거나 처리할 수 있습니다.
for df in fs_local_files:
    print(df)
```
