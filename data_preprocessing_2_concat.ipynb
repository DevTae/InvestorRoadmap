```python
# 가격 데이터 불러오기

import os
import pandas as pd

# 파일 위치를 변수로 지정
directory_path = r"C:\Users\user\OneDrive\바탕 화면\db 공모전\price_stat_datas (2)" # 여기에 실제 디렉토리 경로를 입력하세요.

# 해당 디렉토리의 파일 목록 가져오기
file_list = os.listdir(directory_path)

# CSV 파일들의 이름을 담을 리스트 초기화
price_files = []

# 각 파일에 대해 확인
for file in file_list:
    if file.endswith('.csv') and '2022_4' in file:
        price_files.append(file)

# CSV 파일 목록 출력
print(f"{directory_path} 디렉토리에 있는 CSV 파일 목록:")
print(price_files)



```

```python
# 재무 데이터 가져오기 

# 파일 위치를 변수로 지정
directory_path = r"C:\Users\user\OneDrive\바탕 화면\db 공모전\data\2022\4\pre_data" # 여기에 실제 디렉토리 경로를 입력하세요.

# 해당 디렉토리의 파일 목록 가져오기
file_list = os.listdir(directory_path)

# CSV 파일들의 이름을 담을 리스트 초기화
ifrs_files = []

# 각 파일에 대해 확인
for file in file_list:
    if file.endswith('.csv'):
        ifrs_files.append(file)

# CSV 파일 목록 출력
print(f"{directory_path} 디렉토리에 있는 CSV 파일 목록:")
print(ifrs_files)
```

```python
for i in price_files :
    for j in ifrs_files :
        ifrs_file_name = j[:6]
        if ifrs_file_name in i :
            price_file_path = r"C:\Users\user\OneDrive\바탕 화면\db 공모전\price_stat_datas (2)\{}".format(i)
            ifrs_file_path = r"C:\Users\user\OneDrive\바탕 화면\db 공모전\data\2022\4\pre_data\{}".format(j)
            price_file = pd.read_csv(price_file_path, encoding='utf-8')
            ifrs_file = pd.read_csv(ifrs_file_path, encoding='euc-kr')
            total_data = pd.concat([ifrs_file, price_file], axis= 1, ignore_index= False)
            total_data_path = r'C:\Users\user\OneDrive\바탕 화면\db 공모전\data\2022\4_concat\{}'.format(j)
            total_data.to_csv(total_data_path, encoding='euc-kr')
        else : pass
```
